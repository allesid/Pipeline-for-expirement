```python
# divide 'volume' and'segmentations' dirs and create links to sample and targets files
# in train, val and test dirs 
# data_path
#    \- train
#    \    \ - samples
#    \    \ - targets
#    \
#    \- val
#    \    \ - samples
#    \    \ - targets
#    \
#    \- test
#         \ - samples
#         \ - targets

import os

data_path = "/run/media/alex/PartOfDiskWithWin7/LEARN/Longevity/Liver/datasets/kaggle.com_andrewmvd_liver-tumor-segmentation/"

sample_targets = ['segmentations','volume']
st_dirs = ['targets','samples']
samples_init = sorted(os.listdir(os.path.join(data_path, 'volume')))
targets_init = sorted(os.listdir(os.path.join(data_path, 'segmentations')))
st = zip(targets_init, samples_init)
print(st, next(st))
train_dir = 'train'
val_dir = 'val'
test_dir = 'test'

for dir_name in [train_dir, val_dir, test_dir]:
    for st_dn in st_dirs:
        os.makedirs(os.path.join(data_path, dir_name, st_dn), exist_ok=True)


for i, file_name in enumerate(st):
    print(i, file_name)
    if i % 6 == 0:
        dest_dir_s = os.path.join(data_path, val_dir, st_dirs[1])
        dest_dir_t = os.path.join(data_path, val_dir, st_dirs[0])
    elif i % 6 == 1:
        dest_dir_s = os.path.join(data_path, test_dir, st_dirs[1])
        dest_dir_t = os.path.join(data_path, test_dir, st_dirs[0])
    else:
        dest_dir_s = os.path.join(data_path, train_dir, st_dirs[1])
        dest_dir_t = os.path.join(data_path, train_dir, st_dirs[0])
        # print(file_name[j], '\n', dest_dir, '\n',
            #   os.path.join(data_path, sample_targets[j]))
    os.symlink(os.path.join(data_path, sample_targets[0], file_name[0]),
               os.path.join(dest_dir_t, file_name[0]), dir_fd=None)  # .split('.')[0]
    os.symlink(os.path.join(data_path, sample_targets[1], file_name[1]),
               os.path.join(dest_dir_s, file_name[1]), dir_fd=None)  # .split('.')[0]
```
