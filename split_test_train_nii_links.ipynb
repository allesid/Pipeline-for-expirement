```python
# divide 'volume' and'segmentations' dirs and create links to sample and targets
# in train, val and test dirs

import os

# data_path = "/run/media/alex/PartOfDiskWithWin7/LEARN/Longevity/Liver/datasets/kaggle.com_andrewmvd_liver-tumor-segmentation/"

data_path = "/home/alex/GitLabProjects/research/"
sample_targets = ['segmentations','volume']
samples = sorted(os.listdir(os.path.join(data_path, 'volume')))
targets = sorted(os.listdir(os.path.join(data_path, 'segmentations')))
st = zip(targets, samples)
print(st, next(st))
train_dir = 'nii-train'
val_dir = 'nii-val'
test_dir = 'nii-test'

for dir_name in [train_dir, val_dir, test_dir]:
    os.makedirs(os.path.join(data_path, dir_name), exist_ok=True)

for i, file_name in enumerate(st):
    print(i, file_name)
    if i % 6 == 0:
        dest_dir = os.path.join(data_path, val_dir)
    elif i % 6 == 1:
        dest_dir = os.path.join(data_path, test_dir)
    else:
        dest_dir = os.path.join(data_path, train_dir)
    for j in [0,1]:
        print(file_name[j], '\n', dest_dir, '\n',
              os.path.join(data_path, sample_targets[j]))
        os.symlink(os.path.join(data_path, sample_targets[j], file_name[j]),
                   os.path.join(dest_dir, file_name[j]), dir_fd=None)  # .split('.')[0]
```
