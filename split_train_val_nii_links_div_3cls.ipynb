```python
# %%time
# divide 'volume' and'segmentations' dirs and create links to sample and targets files
# in train and val dirs and 3 classes
# data_path
#    \- train
#    \    \ - disease_liver
#    \    \ - liver
#    \    \ - no liver
#    \
#    \- val
#    \    \ - decease_liver
#    \    \ - liver
#    \    \ - no liver

import os
from torchvision.io import read_image, ImageReadMode
# import torch
import numpy as np
import shutil
# from matplotlib.image import imread
# import cv2
data_path = "/home/alex/GitLabProjects/research/"

sample_targets = ['jpg_nii_segmentation', 'jpg_nii_volume']
st_dirs = ['disease_liver', 'liver', 'no_liver']
samples_init = sorted(os.listdir(os.path.join(data_path, sample_targets[1])))
targets_init = sorted(os.listdir(os.path.join(data_path, sample_targets[0])))
st = zip(targets_init, samples_init)
# print(st, next(st))
train_dir = 'nii_train'
val_dir = 'nii_val'

for dir_name in [train_dir, val_dir]:
    for st_dn in st_dirs:
        os.makedirs(os.path.join(data_path, dir_name, st_dn), exist_ok=True)

for i, file_name in enumerate(st):
    print(i, file_name)
    if file_name[0].endswith((".jpg", ".png", ".jpeg")):
        # img = imread(os.path.join(data_path, sample_targets[0], file_name[0]))
        img = read_image(os.path.join(
            data_path, sample_targets[0], file_name[0]), mode=ImageReadMode.GRAY).numpy()
        # img = cv2.imread(os.path.join(
        #     data_path, sample_targets[0], file_name[0]), cv2.IMREAD_GRAYSCALE)
    else:
        print("Bad extention")
        continue
    # print(np.unique(img))
    # print(img[img == 0].shape)
    if any(img[img > 200]):
        if any(img[(img < 200) & (img > 100)]):
            dd_dir = st_dirs[0]
            # print("disease_liver")
        else:
            dd_dir = st_dirs[1]
            # print("liver")
    else:
        dd_dir = st_dirs[2]
        # print("no_liver")
    if int(file_name[0].split('-')[1]) % 6 == 0:
        dest_dir = os.path.join(data_path, val_dir, dd_dir)
    else:
        dest_dir = os.path.join(data_path, train_dir, dd_dir)
        # print(file_name[j], '\n', dest_dir, '\n',
        #   os.path.join(data_path, sample_targets[j]))
    shutil.copy(os.path.join(data_path, sample_targets[1], file_name[1]),
                os.path.join(dest_dir, file_name[1]))
    # os.symlink(os.path.join(data_path, sample_targets[1], file_name[1]),
    #            os.path.join(dest_dir, file_name[1]), dir_fd=None)  # .split('.')[0]
    if dd_dir != st_dirs[2]:
        shutil.copy(os.path.join(data_path, sample_targets[0], file_name[0]),
                    os.path.join(dest_dir, file_name[0]))
    # os.symlink(os.path.join(data_path, sample_targets[0], file_name[0]),
    #             os.path.join(dest_dir, file_name[0]), dir_fd=None)  # .split('.')[0]
```
